```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

```python
adj = 0
a = 1 - adj
b = 42 - adj
x = np.arange (0.01, 1, 0.01)
y = beta.ppf(x,a,b)
plt.plot(x,y)
```

```python
for x in [.1, .2, .5, .8, .9, .95, .98, .99, .999]:
  print(f"{x:.3f}: {beta.ppf(x, a, b)}")
```

```python
a = 27
b = 57
x = np.arange (0.01, 1, 0.01)
y = beta.pdf(x,a,b)
plt.plot(x,y)
```
